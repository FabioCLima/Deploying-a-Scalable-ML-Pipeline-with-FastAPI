## Verificando se há falso nulos

```python
print((census_df == '?').sum())
```

```python
# Preenche os nulos com 'Unknown' - Rápido e seguro para a API
cols_with_nan = ['workclass', 'occupation', 'native-country']
census_df[cols_with_nan] = census_df[cols_with_nan].fillna('Unknown')
```

```python
import matplotlib.pyplot as plt

for col in colunas_numericas:
    census_df[col].hist()
    plt.title(f"Distribuição de {col}")
    plt.show()
```
